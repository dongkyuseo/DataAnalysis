### 반드시 런타임 다시시작 해야 함

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
# %matplotlib inline 과거에는 주피터노트북을 사용하며 그래프를 보기위해 사용했던 매직 명령어, 지금은 사용안해도 됨
```

```python
plt.title("라인 플롯")
plt.plot([1, 4, 9, 16])     # y = (x + 1) ** 2
plt.grid()                  # 격자선
plt.show()                  # 그래프 시각화
```

```python
plt.title("라인플롯")
xs = np.linspace(0, 3, 301)
ys = (xs + 1) ** 2
plt.plot(xs, ys)
plt.grid()
plt.show()
```

### 스타일 지정

```python
# 색깔(r 빨강), 마커(s 네모) 점: 아래의 네모 점, -- 선스타일
plt.title("'rs--' 스타일의 plot ")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--')   #x, y, rs-- : 
plt.grid()
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b",
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("스타일 적용 예")
plt.grid()
plt.show()
```

### 그림 범위 지정

```python
plt.title("x축, y축의 범위 설정")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50)   #x축 범위지정
plt.ylim(-2, 20) #y축 범위지정
plt.grid()
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.grid()
plt.show()
```

```python
# 틱 라벨 문자열에는 $$ 사이에 LaTeX 수학 문자식을 넣을 수도 있다.
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("LaTeX, 문자열로 tick label 정의")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.grid()
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

```python
# 코드를 직관적으로 라인을 구분해서 하는게 보기 쉬움
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--')
plt.plot(t, 0.5 * t**2, 'bs:')
plt.plot(t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

### 범례(Legend)

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, label="cosine")    #범례를 주려면 laber 값을 줘야 값이 같이 출력됨
plt.plot(X, S, label='shine')
plt.legend(loc='lower center')
plt.grid()
plt.show()
```

### x축, y축 라벨, 타이틀

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.grid()
plt.show()
```
