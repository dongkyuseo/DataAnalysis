```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
```

```python
import seaborn as sns
iris = sns.load_dataset('iris') # 붓꽃 데이터
titanic = sns.load_dataset('titanic') # 타이타닉호 데이터
tips = sns.load_dataset('tips') # 팁 데이터
flights = sns.load_dataset('flights') # 여객운송 데이터
```

## Line Plot

```python
np.random.seed(0)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2018', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```python
df1.plot()
plt.title('Pandas의 plot() 메서드 사용 예')
plt.xlabel('일자')
plt.ylabel('Data')
plt.grid()
plt.show()
```

## Bar chart

```python
iris.sepal_length[:20].plot(kind='bar', rot=0) # color=plt.get_cmap('hot')(np.linspace(0.8,0.2,20) 컬러를 넣기위해 해야 한는 수식, 복잡함
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
iris.sepal_length[:20].plot(kind='bar', rot=0, color=plt.get_cmap('RdBu')(np.linspace(0.8,0.2,20)))
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
iris.sepal_length[:20].sort_values().plot(kind='barh', rot=0, # 내림차순: sort_values(), 오름차순 : sort_values(ascending=False)
                            color=plt.get_cmap('RdBu')(np.linspace(0.8,0.2,20)))
plt.title("꽃받침의 길이 시각화")
plt.ylabel("Data")
plt.xlabel("꽃받침의 길이")
plt.show()
```

```python
iris[:5].plot.bar(rot=0)
plt.title("Iris 데이터의 Bar Plot")
plt.xlabel("Data")
plt.ylabel("각 Feature의 값")
plt.ylim(0, 7)    # 범례 중간으로 이동
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = "feature"
df2
```

```python
df2.plot.bar(rot=0)
plt.ylim(0,9)
plt.title('각 종의feature별 평균')
plt.show()
```

```python
df2.T.plot.bar(rot=0)
plt.title('각 feature의 종별 평균')
plt.show()
```

## Pie chart

```python
df3 = titanic['class'].value_counts()
df3.plot.pie(autopct='%.2f%%')
plt.title("선실별 승객 수 비율")
plt.axis('equal')
plt.show()
```

## Histogram

```python
iris.plot.hist()
plt.title("각 Feature 값들의 빈도수 Histogram")
plt.xlabel("데이터 값")
plt.show()
```

## Box plot

```python
iris.plot.box()
plt.title("각 Feature 값들의 빈도수에 대한 Box Plot")
plt.xlabel("Feature")
plt.ylabel("데이터 값")
plt.show()
```

```python
iris.boxplot(by='species')
plt.tight_layout(pad=2, h_pad=1)
plt.title("각 Feature의 종 별 데이터에 대한 Box Plot")
plt.show()
```
