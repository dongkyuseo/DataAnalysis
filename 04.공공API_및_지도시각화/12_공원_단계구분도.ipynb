# 단계구분도

```python
import pandas as pd
import folium
import json
```

```python
from google.colab import files
uploaded = files.upload()
filename = list(uploaded.keys())[0]
```

```python
df = pd.read_csv(filename)
df.head()
```

```python
uploaded = files.upload()
gu_geodata = list(uploaded.keys())[0]
```

```python
geo_data = json.load(open(gu_geodata))
```

- 자치구별 공원면적

```python
center = [37.581, 126.986]    # 서울 중심부 위도, 경도
map = folium.Map(location=center, zoom_start=11, tiles='Stamen Toner')

folium.TileLayer('cartodbpositron').add_to(map)
folium.Choropleth(
    geo_data=geo_data,
    data = df,
    columns = ('지역','공원면적'),
    key_on = 'feature.id',    # geo_data 의 지역명
    fill_color = 'BuPu',
    legend_name = '공원면적'
).add_to(map)
title = '<h3 align="center" style="font-size:20px">서울시 구별 공원면적</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```

```python
uploaded = files.upload()
filename = list(uploaded.keys())[0]
```

```python
park = pd.read_csv(filename)
park.head()
```

```python
center = [37.581, 126.986]    # 서울 중심부 위도, 경도
map = folium.Map(location=center, zoom_start=11, tiles='Stamen Toner')

folium.TileLayer('cartodbpositron').add_to(map)
folium.Choropleth(
    geo_data=geo_data,
    data = df,
    columns = ('지역','공원면적'),
    key_on = 'feature.id',    # geo_data 의 지역명
    fill_color = 'BuPu',
    legend_name = '공원면적'
).add_to(map)

for i in park.index:
  folium.CircleMarker(
      radius=int(park.크기[i]),                                   # int64는 처리 불가
      location=[park.위도[i], park.경도[i]],
      popup=folium.Popup(f'{park.공원명[i]}', max_width=300),   # popup 누르면 나오는 문구, max_widht 팝업창 넓이
      tooltip=f'{park.공원명[i]},({park.면적[i]}㎡)',   # 커서온시 문구
      color='green',    # 마커 색
      fill=True,   # 마커 채우기
      fill_color='green'
  ).add_to(map)

title = '<h3 align="center" style="font-size:20px">서울시 구별 공원면적, 공원위치 및 크기</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```
