```python
import pandas as pd
raw_gu_df = pd.read_csv('서울시_구별_인구밀도.txt', delimiter = '\t')
raw_gu_df
```

```python
raw_gu_df['구면적'] = raw_gu_df.면적.astype(int).apply(lambda x: int(x) * 1000000)
raw_gu_df
```

```python
import re
raw_gu_df['인구수'] = raw_gu_df.인구.apply(lambda x: int(re.sub('[,]','',x)))
raw_gu_df
```

```python
gu_df =[]
gu_df = raw_gu_df.loc[:,['지역','구면적','인구수']]
gu_df.set_index('지역', inplace=True)
gu_df
```

```python
import pandas as pd
raw_df = pd.read_csv(filename)
raw_df
```

```python
import numpy as np
```

```python
#raw_df.면적.value_counts()

raw_df.공원명.count()
```

```python
df = raw_df.groupby(raw_df.지역)[["면적"]].sum()
df
```

```python
df["공원수"] = raw_df.groupby(raw_df.지역)[["공원명"]].count()
df
```

```python
df = df.drop(index='0', axis=0)
df
```

```python
gu_df
```

```python
df1 = pd.merge(df, gu_df, on='지역')
df1
```

```python
df1 = df1.rename(columns={
    '면적':'공원면적',
    '공원수':'공원수',
    '구면적':'구면적',
    '인구수':'인구수'
})
df1
```

```python
df1.to_csv('서울시구별공원면적공원수구면적인구수.csv')
```

```python
df = pd.read_csv('서울시구별공원면적공원수구면적인구수.csv')
df.tail()
```

```python
import seaborn as sns
```

```python
df2 = df[['지역','공원면적']].sort_values(by='공원면적', ascending=False)
plt.figure(figsize=(12,8))
plt.title('서울시 구별 공원정보')
sns.barplot(y='지역', x='공원면적', data=df2)

plt.show()
```

```python
df['인당공원면적'] = df.공원면적 / df.인구수
df
```

```python
df['공원면적비율'] = df.공원면적 / df.구면적 * 100
df
```

```python
df2 = df[['지역','공원면적비율']].sort_values(by='공원면적비율', ascending=False)
plt.figure(figsize=(12,8))
plt.title('서울시 구별 공원면적비율', size=15)
sns.barplot(y='지역', x='공원면적비율', data=df2)

plt.show()
```

```python
df2 = df[['지역','인당공원면적']].sort_values(by='인당공원면적', ascending=False)
plt.figure(figsize=(12,8))
plt.title('서울시 구별 인당공원면적비율', size=15)
sns.barplot(y='지역', x='인당공원면적', data=df2)

plt.show()
```
