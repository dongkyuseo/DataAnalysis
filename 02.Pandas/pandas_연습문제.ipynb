```python
# a
titanic['category1'] = titanic.apply(lambda x : x.sex if x.age >= 20 else 'child', axis=1)
titanic.tail()
```

```python
#b
titanic.age.fillna(titanic.age.mean(), inplace=True)    # inplace 를 False 에서 True 로 바꿔야 보임
titanic.tail()
```

```python
#c
titanic.pivot_table('survived', 'sex')
```

```python
titanic.pivot_table('survived', 'class')
```

```python
titanic.pivot_table('survived', 'embark_town')
```

```python
titanic.pivot_table('survived', 'sex', ['class', 'embark_town'])
```

```python
# d
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년 ", "중년 ", "장년 ", "노년"]
```

```python
titanic['labels'] = pd.cut(titanic.age, bins, labels=labels)
titanic.tail()
```

```python
# e
titanic['abc'] = pd.qcut(titanic.age, 3, labels=['a','b','c'])
titanic.tail()
```

```python
titanic.pivot_table('survived', 'sex', 'abc')
```

```python
import seaborn as sns
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
# a
mpg['hp_per_cc'] = mpg.horsepower / mpg.displacement
```

```python
mpg.head()
```

```python
# b
mpg['manufacturer'] = mpg.apply(lambda x : x for x in [mpg.name.split()][0])
mpg.head()
```

```python
mpg['model']
```

```python
# c
mpg.pivot_table('mpg', 'cylinders' )
```

```python
# d
mpg.pivot_table('hp_per_cc', 'origin')
```
